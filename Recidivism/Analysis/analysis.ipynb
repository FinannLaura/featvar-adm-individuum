### Descriptive statistics 

```python
# Function to compute mean, std, and median for each row
def compute_row_stats(file):

    numeric_cols = file.columns[1:] 

    file[numeric_cols] = file[numeric_cols].apply(pd.to_numeric, errors='coerce')

    row_stds = file[numeric_cols].std(axis=1, skipna=True)

    row_means = file[numeric_cols].mean(axis=1, skipna=True)

    row_medians = file[numeric_cols].median(axis=1, skipna=True)
    
    stats_df = pd.DataFrame({
        'ID': file['ID'],  
        'mean': row_means,
        'std': row_stds,
        'median': row_medians
    })
    return stats_df
```

```python
# Dataframes with mean, std, and median for 2L, 7L, 10L, and combined_L Model
two_L = compute_row_stats(df_2L)
seven_L =  compute_row_stats(df_7L)
ten_L = compute_row_stats(df_10L)
combined_L = compute_row_stats(combined_L)
```

```python
# Function to combine the standard deviations of the models into a single DataFrame
def combine_std(file1, file2, file3, file4):

    combined_std = pd.concat([file1['std'], file2['std'], file3['std'], file4['std']], axis=1)
    combined_std.columns = ['std_2', 'std_7', 'std_10', 'std_combined']
    return combined_std
```

### Model Multiplicity

```python
# Function to plot histograms
def plot_histogram(data, title, color, ax, bins, x_limit, y_limit):

    sns.histplot(data, kde=False, color=color, ax=ax, element='bars', bins=bins)
    ax.set_title(title)
    ax.set_xlabel('Standard Deviation')
    ax.set_ylabel('Frequency')
    ax.set_xlim([0, x_limit])
    ax.set_ylim([0, y_limit])

# Function to calculate the limits for the histograms
def calculate_limits(combined_std):
    max_std_value = max(combined_std['std_2'].max(), combined_std['std_7'].max(), 
                        combined_std['std_10'].max(), combined_std['std_combined'].max())
    #print(max_std_value)

    bins = np.arange(0, max_std_value + 0.01, 0.01)

    y_limit = 0
    for data in [combined_std['std_2'], combined_std['std_7'], combined_std['std_10'], combined_std['std_combined']]:
        hist, _ = np.histogram(data, bins=bins)
        y_limit = max(y_limit, hist.max())
    
    y_limit *= 1.1
    
    return max_std_value, bins, y_limit

# Function to plot model multiplicity for each model type
def plot_models(combined_std, model_name, axes, bins, max_std_value, y_limit):
    features = ['2', '7', '10', 'combined']
    colors = ['gold', 'olive', 'skyblue', 'teal']
    titles = [f'{model_name} Model with {feat} Features' for feat in features]
    
    for i, feat in enumerate(features):
        plot_histogram(combined_std[f'std_{feat}'], titles[i], colors[i], axes[i], bins, max_std_value, y_limit)


model_data = {
    'LR': combined_std_L,
    'RF': combined_std_R,
    'LR Binary': combined_std_BL,
    'RF Binary': combined_std_BR

}

for model_name, combined_std in model_data.items():
    max_std_value, bins, y_limit = calculate_limits(combined_std)
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    axes = axes.flatten() 
    
    plot_models(combined_std, model_name, axes, bins, max_std_value, y_limit)
    
    plt.tight_layout()
    plt.show() 
```

### Mean Distribution

```python
# Function to calculate the limits for the histograms
def calculate_max_y(data):

    return sns.histplot(data['mean'], bins=50, stat="count").get_ylim()[1]


# Function to plot the distribution of means for a single dataset
def plot_mean_distribution(df, title, color, max_y):
  
    fig, ax = plt.subplots(figsize=(8, 6))

    sns.histplot(df['mean'], kde=False, color=color, ax=ax, element='bars', 
                 bins=50, stat="count") 
    
    ax.set_title(title, fontsize=14, fontweight='normal', color='#333333')
    ax.set_xlabel('Mean for each data point across all 1000 models', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    
    ax.set_xlim(left=0, right=1)  
    ax.set_ylim(bottom=0, top=max_y) 

    ax.set_facecolor('#f7f7f7')
    
    sns.despine(ax=ax, top=True, right=True)
    
    plt.tight_layout()
    return fig, ax
```

### CCDF

```python
# Function to compute and plot CCDF
def plot_ccdf(data, column, title, color, threshold):

    data = data.dropna(subset=[column])
    #print(data.head())
    sorted_data = data[column].sort_values(ascending=False)

    ccdf = sorted_data.reset_index(drop=True).reset_index()
    ccdf.columns = ['Rank', 'Relative difference']
    ccdf['Percentage'] = (ccdf['Rank'] / len(ccdf))
    

    percentage = (data[column] >= threshold).mean() * 100

    # Plot CCDF
    plt.figure(figsize=(8, 6))
    plt.plot(ccdf['Relative difference'], ccdf['Percentage'], color=color)
    plt.axvline(x=threshold, color='pink', linestyle='--',label=f'{threshold} threshold', ymin=0, ymax=1)
    plt.xlabel('Relative binary difference')
    plt.ylabel('Cumulative Probability')  
    # plt.xlim( 0, max(ccdf['Relative difference']))
    plt.ylim(0, 1)
    plt.title(title)
    plt.legend()
    plt.show()
    print(f'Percentage of points with at least {threshold} absolute difference in means: {percentage:.2f}%\n')
```

```python
# Plot CCDF for 2 vs. 7 and 2 vs. 10 for LR Models
plot_ccdf(abs_mean_diffs_L, 'abs_diff_2_7', 'CCDF of Absolute Differences: 2 vs. 7 Feature LR Models', 'gold', 0.05)
plot_ccdf(abs_mean_diffs_L, 'abs_diff_2_10', 'CCDF of Absolute Differences: 2 vs. 10 Feature LR Models', 'olive', 0.1)

# Plot CCDF for 2 vs. 7 and 2 vs. 10 for RF Models
plot_ccdf(abs_mean_diffs_R, 'abs_diff_2_7', 'CCDF of Absolute Differences: 2 vs. 7 Feature RF Models', 'gold', 0.05)
plot_ccdf(abs_mean_diffs_R, 'abs_diff_2_10', 'CCDF of Absolute Differences: 2 vs. 10 Feature RF Models', 'olive', 0.1)

# Plot CCDF for 2 vs. 7 and 2 vs. 10 for LR Binary Models
plot_ccdf(abs_mean_diffs_BL, 'abs_diff_2_7', 'CCDF of Absolute Differences: 2 vs. 7 Feature LR Binary Models', 'gold', 0.05)
plot_ccdf(abs_mean_diffs_BL, 'abs_diff_2_10', 'CCDF of Absolute Differences: 2 vs. 10 Feature LR Binary Models', 'olive', 0.1)

# Plot CCDF for 2 vs. 7 and 2 vs. 10 for RF Binary Models
plot_ccdf(abs_mean_diffs_BR, 'abs_diff_2_7', 'CCDF of Absolute Differences: 2 vs. 7 Feature RF Binary Models', 'gold', 0.05)
plot_ccdf(abs_mean_diffs_BR, 'abs_diff_2_10', 'CCDF of Absolute Differences: 2 vs. 10 Feature RF Binary Models', 'olive', 0.1)
```

### Relative Binary Difference

```python
# Function to compare binary models by calculating the relative differences 
def compare_binary_models(df_1, df_2, title, feature_type_1='2', feature_type_2='7'):

    # Remove the first column
    df_1_binary = df_1.drop(df_1.columns[0], axis=1)
    df_2_binary = df_2.drop(df_2.columns[0], axis=1)

    # Calculate the count of non-NaN values per row
    df_1_count = df_1_binary.count(axis=1)
    df_2_count = df_2_binary.count(axis=1)

    # Add the count as a new column to the DataFrames
    df_1_binary = pd.concat([df_1_binary, df_1_count.rename('count')], axis=1)
    df_2_binary = pd.concat([df_2_binary, df_2_count.rename('count')], axis=1)

    # Save the count column to a new DataFrame
    df_1_binary_count = df_1_binary['count']
    df_2_binary_count = df_2_binary['count']

    # Shorten column names by removing the first two characters 
    df_1_binary.columns = [col[2:] for col in df_1_binary.columns]
    df_2_binary.columns = [col[2:] for col in df_2_binary.columns]

    # if first character is an L or R, remove it
    df_1_binary.columns = [col[1:] if col[0] in ['L', 'R'] else col for col in df_1_binary.columns]
    df_2_binary.columns = [col[1:] if col[0] in ['L', 'R'] else col for col in df_2_binary.columns]


    # Calculate the absolute difference between corresponding columns
    df_diff = np.abs(df_1_binary - df_2_binary).sum(axis=1)

    # Calculate the relative difference
    if df_1_binary_count.equals(df_2_binary_count):
        df_diff = df_diff.div(df_1_binary_count, axis=0).round(2)
        diff_mean = df_diff.mean()
    else:
        raise ValueError("Counts are not equal")

    # Plot the histogram
    plt.figure()
    counts, bins, patches = plt.hist(df_diff, bins=20, color='blue', alpha=0.7, density=False)
    counts_percentage = (counts / counts.sum())
    plt.clf()  
    plt.bar(bins[:-1], counts_percentage, width=np.diff(bins), align="edge", color='#FF5733', alpha=0.7)
    plt.xlabel('Relative Difference')
    plt.ylabel('Percentage of Predictions')
    plt.title(f'Distribution of Relative Binary Differences: {feature_type_1} vs. {feature_type_2} {title}')
    plt.axvline(diff_mean, color='limegreen', linestyle='dashed', linewidth=2, label=f'Mean: {diff_mean:.2f}')
    plt.legend()
    plt.ylim(0, 1)
    plt.show()

    return diff_mean
```

```python
# Relative Binary Differences: 2 vs. 7 Feature LR Models
mean_diff_lr = compare_binary_models(df_2BR, df_7BR, title='LR Feature Models', feature_type_1='2', feature_type_2='7')

# Relative Binary Differences: 2 vs. 10 Feature LR Models
mean_diff_lr_2 = compare_binary_models(df_2BR, df_10BR, title='LR Feature Models', feature_type_1='2', feature_type_2='10')

# Relative Binary Differences: 2 vs. 7 Feature RF Models
mean_diff_rf_10 = compare_binary_models(df_2BR, df_7BR, title='RF Feature Models', feature_type_1='2', feature_type_2='7')

# Relative Binary Differences: 2 vs. 10 Feature RF Models
mean_diff_rf_10 = compare_binary_models(df_2BR, df_10BR, title='RF Feature Models', feature_type_1='2', feature_type_2='10')

# Relative Binary Differences: 2 vs. 7 Feature LR Models
mean_diff_lr = compare_binary_models(df_2L, df_7L, title='LR Feature Models non binary', feature_type_1='2', feature_type_2='7')
```
